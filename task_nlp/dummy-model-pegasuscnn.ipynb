```python
n = 3000
print(len(text))
substrings = [text[i:i+n] for i in range(0, len(text), n)]
print(substrings)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/pegasus-cnn_dailymail")

model = AutoModelForSeq2SeqLM.from_pretrained("google/pegasus-cnn_dailymail")
```

```python
from datetime import datetime

def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
result = []

start_time = timer(None)
for string in substrings:
    input_ids = tokenizer.encode(string, return_tensors="pt", truncation=True)
    generated_sequence = model.generate(
        input_ids=input_ids, 
        num_beams=10, 
        max_new_tokens=300, 
        min_length=50, 
        top_k=30,
    )
    output_text = tokenizer.decode(generated_sequence.squeeze(), skip_special_tokens=True)
    result.append(output_text)
    print(output_text)
timer(start_time)
```

```python
overall_output = ' '.join(result)
print(len(overall_output))
overall_output
```

```python
start_time = timer(None)
input_ids = tokenizer.encode(overall_output, return_tensors="pt", truncation=True)
generated_sequence = model.generate(
    input_ids=input_ids, 
    num_beams=10, 
    max_new_tokens=500, 
    min_length=250, 
    top_k=30,
)
output_text = tokenizer.decode(generated_sequence.squeeze(), skip_special_tokens=True)
print(output_text)
timer(start_time)
```
